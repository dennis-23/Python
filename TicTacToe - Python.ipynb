Python script for TicTacToe.


```python
from IPython.display import clear_output

def display_board(board): #Display Board on each iteration
    clear_output()
    print(board[7]+" | "+board[8]+" | "+board[9])
    print("--|---|--")
    print(board[4]+" | "+board[5]+" | "+board[6])
    print("--|---|--")
    print(board[1]+" | "+board[2]+" | "+board[3])
    
    
def player_input():
    #Func to choose who is going to be X and O
    marker = ''
    
    while not (marker == 'X' or marker == 'O'):
        marker = input('Player 1: Do you want to be X or O? ').upper()

    if marker == 'X':
        return ('X', 'O')
    else:
        return ('O', 'X')
    

def place_marker(board, mark, position):
    #Check if the position chosen is valid based on range and availability
    while position not in range(1,10):
        position=int(input("Enter position 1-9:"))
    
    board[position]=mark
    
    print("The updated board is: ")
    print('\n')
  #  print(board[9]+" | "+board[8]+" | "+board[7])
  #  print("--|---|--")
  #  print(board[4]+" | "+board[5]+" | "+board[6])
  #  print("--|---|--")
  #  print(board[1]+" | "+board[2]+" | "+board[3])
  #  print("--|---|--")   

def win_check(board, mark):
    #Win condition check that must be run after every move
    if board[5]==board[4]==board[6]==mark or board[1]==board[4]==board[7]==mark or board[7]==board[8]==board[9]==mark or board[9]==board[6]==board[3]==mark or board[1]==board[2]==board[3]==mark or board[7]==board[5]==board[3]==mark or board[9]==board[5]==board[1]==mark:
                        return True

import random

def choose_first():
    #Choosing the player who goes first - Random basis
    if random.randint(1,2)==1:
        return "Player 1"
    else:
        return "Player 2"
    
def space_check(board, position):
    #Check for empty space in chosen position
    return board[position]==" "
        
def full_board_check(board):
    #Check if the whole board has been filled
    for i in range(1,10):
        #3if space_check(board,i)
        if board[i]!=" ":
            return False
    return True

def player_choice(board):
    #Func to take the player's position choice
    position=0
    while position not in range(1,10) or not space_check(board,position):
        position=int(input("Enter Next position (1-9): "))
    return position


def replay():
    #Replay choice at the end of the game
    play_again=""
    while play_again!="Y" and play_again!="N":
        play_again=input("Do you want to play again?(Y or N): ")
    if play_again=="Y":
        return True
    else:
        return False
```

```python
#WORKING!!

print('Welcome to Tic Tac Toe!')

while True:#Keeps the game logic looping until termination
    # Reset the board
    theBoard = [' ']*10 #Board layouy consists of 10 blocks of which 9 will b displayed
    player1, player2=player_input() #Tuple unpacking for Player symbol
    turn = choose_first() 
    print(turn + ' will go first.')
    
    play_game=input('Are you ready to play? Enter Yes or No.')
    
    if play_game=="Yes":
        game_on=True #game_on condition determines if the game is to be continued or terminated
    else:
        game_on=False

    while game_on:
        if turn=="Player 1": 
            # Player1's turn.
            
            display_board(theBoard)
            position=player_choice(theBoard)
            place_marker(theBoard, player1, position)

            #After each insertion the board is checked for a win or draw condition before the control goes to the opponent
            if win_check(theBoard, player1): 
                display_board(theBoard)
                print("\nCongratulations! Player 1 has won the game!")
                game_on=False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print("\nThe game is a draw!")
                    break
                else:
                    turn="Player 2"

        else:
            # Player2's turn.
            
            display_board(theBoard)
            position=player_choice(theBoard)
            place_marker(theBoard, player2, position)

            if win_check(theBoard, player2):
                display_board(theBoard)
                print('\nCongratulations! Player 2 has won the game!')
                game_on=False
            else:
                if full_board_check(theBoard):
                    display_board(theBoard)
                    print('\nThe game is a draw!')
                    break
                else:
                    turn='Player 1'

    if replay()==False:
        print("\nThank you for playing!")
        break
```
